```python
import pandas as pd
import numpy as np

# col_A = ['A', 'B', 'C']
# data = [[1, 2, 3], [11, 12, 13], [21, np.nan, 23]]
data = {'A': ['a1', 'a1', 'a2'],
        'B': ['b1', 'b2', 'b3'], 
        'C': ['no_affect', np.nan, 'no_affect'],
        'D': ['stay', np.nan, 'to_change']}
A = pd.DataFrame(data)
display(A)

data = {'A': ['a1', 'a1', 'a2', 'a2'],
        'B': ['b1', 'b2', 'b3', 'b4'], 
        'D': ['stay', 'updated', 'changed', np.nan], 
        'E': [31, 32, np.nan, 34]}
B = pd.DataFrame(data)
display(B)
```

```python
C = A.merge(B, on=['A', 'B'], how='outer', suffixes=('_x', ''))

str_a = 'somestr_x'
print(str_a[-2:]=='_x')
# fillna(_x)

print(C)
```

```python
A.update(B)
display(A)
col_to_join = B.columns.difference(A.columns)
C = A.merge(B[col_to_join], how='left', left_index=True, right_index=True)
display(C)
```

```python
D = A.merge(C, left_on=['A', 'B'], right_on=['A', 'B'], how='left')
print(D)
```
