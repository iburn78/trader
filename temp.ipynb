```python
import pandas as pd
import numpy as np

# col_A = ['A', 'B', 'C']
# data = [[1, 2, 3], [11, 12, 13], [21, np.nan, 23]]
data = {'A': [1, 2, 3],
        'B': [11, 12, 13], 
        'C': [21, np.nan, 23]}
A = pd.DataFrame(data)
display(A)

col_B = ['B', 'C', 'D']
data = [[np.nan, np.nan, 113], [21, 22, 23], [31, 32, np.nan]]
data = {'B': [np.nan, np.nan, 113],
        'C': [21, 22, 23], 
        'D': [31, 32, np.nan]}
B = pd.DataFrame(data, columns=col_B)
display(B)
```

```python
A.update(B)
display(A)
col_to_join = B.columns.difference(A.columns)
C = A.merge(B[col_to_join], how='left', left_index=True, right_index=True)
display(C)
```
